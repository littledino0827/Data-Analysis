```python
data = data.drop_duplicates()
```

```python
data = data.drop(columns = 'Not_Useful_Column')
data.head()
```

```python
data['Last_Name'] = data['Last_Name'].str.replace(r'[/_.]', '', regex=True)
```

```python
data['Phone_Number'] = data['Phone_Number'].astype(str).str.replace(r'[^A-Za-z0-9]', '', regex=True)
data['Phone_Number'] = data['Phone_Number'].str.replace(r'[nanN]', '', regex=True)
data['Phone_Number'] = data['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:9])
data['Phone_Number'] = data['Phone_Number'].str.replace('--', '', regex=True)
```

```python
data[['Street_Add','State','zipcode']] = data['Address'].str.split(',', n=2, expand = True)
data.drop(columns = 'Address')
```

```python
data['Paying Customer'] = data['Paying Customer'].str.replace('Yes','Y')
data['Paying Customer'] = data['Paying Customer'].str.replace('No','N')
data['Paying Customer'] = data['Paying Customer'].str.replace('N/a','')

data['Do_Not_Contact'] = data['Do_Not_Contact'].str.replace('Yes','Y')
data['Do_Not_Contact'] = data['Do_Not_Contact'].str.replace('No','N')
data['Do_Not_Contact'] = data['Do_Not_Contact'].fillna('')

data
```
