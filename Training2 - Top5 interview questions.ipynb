# 3. Joins & Concat

```python
# Show all the main data and Used Prices per car
pd.merge(car_main, car_used_prices, on = "index")
```

```python
# Show the Make, Model and Average Used Price Per Type(Make/Model)
pd.merge(car_main, car_used_prices, on = "index").groupby(["Make","Model"])["Used Price"].mean().reset_index()
```

```python
# # Show the Make, Model, Retail Price, Cylinders & Horsepower
pd.merge(car_info[["index","Cylinders","Horsepower"]], car_main[["index","Make","Model","Retail_Price"]], on = "index")
```

```python
# # create a master dataset that combines all 3 tables
car_master = pd.merge(car_info, car_main, on = "index")
car_master = pd.merge(car_master, car_used_prices, on = "index")
car_master.head(3)
```

# 4. Creating new fields

```python
car_master.dtypes
```

```python
# Create a new column that has a "Yes/No" if the car retail price is > than 100k
car_master["high_rtl_price"] = np.where(car_master["Retail_Price"] > 100000, "Yes", "No")
car_master.head(3)
```

```python
# Create a new Torque column with just the number
car_master["Torque No"] = car_master["Torque"].str[:3].astype(float)
car_master.tail(3)
```

```python
# Create a new field which has the Retail_price per torque
car_master["RP_ptorque"] = (car_master["Retail_Price"] / car_master["Torque No"]).round(2)
car_master.head(3)
```

# 5. For Loops

```python
# Add the average retail price per Model in the master table
models = car_master["Model"].unique()
models
```

```python
data = car_master.groupby("Model")["Retail_Price"].mean().reset_index(name = "Avg_rp_Model")
data["Avg_rp_Model"] = data["Avg_rp_Model"].round(2)
data.head(3)
```

```python
car_master = pd.merge(car_master, data, how = "left", on = "Model")
car_master.head(3)
```
